```python
# Berlin Food Delivery ETA Prediction
## Step 8: Export Model for Streamlit App

Goal: Train the final model one more time on the full dataset (not just the 80% 
training split) for maximum performance, then export it for use in a live app.
```

```python
import pandas as pd
import numpy as np
import joblib
import json
from sklearn.ensemble import GradientBoostingRegressor
```

```python
## Loaded Feature engineered data

train = pd.read_csv('train_features.csv')

drop_cols = [
    'ID', 'Delivery_person_ID', 'Order_Date', 'Time_Orderd', 'Time_Order_picked',
    'Restaurant_latitude', 'Restaurant_longitude',
    'Delivery_location_latitude', 'Delivery_location_longitude',
    'Weatherconditions', 'Road_traffic_density'
]

X = train.drop(columns=drop_cols + ['Time_taken(min)'])
y = train['Time_taken(min)']

print(X.shape)
print(X.columns.tolist())
```

```python
## trained final model on ALL data using your best tuned params

final_model = GradientBoostingRegressor(
    n_estimators=100, max_depth=6, learning_rate=0.05,
    subsample=0.7, min_samples_split=2, random_state=42
)
final_model.fit(X, y)
print("Final model trained on full dataset")
```

```python
## saved model + exact feature column order 

joblib.dump(final_model, 'delivery_eta_model.pkl')

with open('feature_columns.json', 'w') as f:
    json.dump(X.columns.tolist(), f)

print("Saved delivery_eta_model.pkl and feature_columns.json")
```
